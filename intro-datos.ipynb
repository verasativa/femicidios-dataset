# Distrbuciones

```python
import numpy as np
import seaborn as sns
%matplotlib inline

# Creamos 100 puntos al "azar"
how_many = 100
data = np.random.rand(how_many)

# Imprimos el promedio
mean = np.mean(data)
display(mean)

# Graficamos las distribución en un histograma
sns.distplot(data)
```

```python
import matplotlib.pyplot as plt

# Creamos una figura con tamaño más grande que el default
fig, ax = plt.subplots(figsize=(14, 8))

# Volvemos a graficar las distribución
plt.title('{} puntos al "azar"'.format(how_many))
null = sns.distplot(data, ax=ax) # pasamos el objeto AxesSubplot a una variable para que no lo imprima
```

```python
from collections import namedtuple
import functions as func

from importlib import reload  # Python 3.4+ only.
reload(func)

# Creamos un named tuple
Distribution = namedtuple('Distribution', 'name data')

# Exploremos las distribuciones que nos ofrece numpy
distributions = [
    Distribution('standard_normal', np.random.standard_normal(size=how_many) + .45),
    Distribution('pareto', np.random.pareto(10, size=how_many) + .38),
    Distribution('uniform', np.random.uniform(size=how_many)),
    Distribution('Bi-modal', func.bi_modal(mode_a=.2, mode_b=.8, sd=.1, size=how_many)),
]

for distribution in distributions:
    # Creamos una figura con tamaño más grande que el default
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Volvemos a graficar las distribución
    title = '{} puntos {}'.format(how_many, distribution.name, np.mean(distribution.data))
    plt.title(title)
    sns.distplot(distribution.data, ax=ax)
    
    # Mostramos el grafico y luego liberamos la memoria (estamos en una iteración)
    display(fig)
    plt.close(fig)
    
    # Imprimimos estimaciones de locación / ubicación
    func.display_locations(distribution.data)
```

### Mi pregunta
# ¿Como es la edad de las victimas?
## Ubicación y promedio

```python
# Creamos una figura con tamaño más grande que el default
fig, ax = plt.subplots(figsize=(14, 8))

# Creamos una mascara tipo: [True, True, False, True] para [23, 45, Null, 32] (ante la duda usa print(var) o display(var))
mask = femicidios.Edad.notnull()
# Seleccionamos filas con la mascara, y luego la columna edad
data = femicidios[mask].Edad

plt.title('Distribución de edades femicidios en Chile')
sns.distplot(data, ax=ax)

mean = np.mean(data)
print('Promedio: {}'.format(mean))
median = np.median(data)
print('Media: {}'.format(median))
```

# ¿Y como se comparan las edades de las victimas con la de los femicidas?

```python
# Creamos un dataframe donde almacenar las 2 columnas
edades = pd.DataFrame()

edades['victima'] = femicidios.Edad
edades['femicida'] = femicidios.Edad_femicida

# Eliminamos los valores nulls
edades = edades.dropna()

# Creamos una figura con tamaño más grande que el default
fig, ax = plt.subplots(figsize=(14, 8))

# Titulo
plt.title('Distribución de edades femicidios en Chile')

# ploteamos edad femicidas y almacenamos el plot axis en la variable "ax" para pasarselo a la siguiente serie a plotear
ax = sns.distplot(edades.femicida, ax=ax, label="Femicidas")

# siendo la ultima seríe a graficar, ya no guardamos el axis
sns.distplot(edades.victima, ax=ax, label="Victimas")

# Mejor presentación seteando titulo, leyenda y etiqueta x (comenta para ver qué desaparece)
plt.title('Distribución de edades femicidios en Chile')
plt.legend()
null_var = ax.set_xlabel('Años') # Para que no retorne el objeto label
```

```python
# Convertimos la serie de objetos fecha, a una serie de fechas (dtype)
fechas = pd.to_datetime(femicidios.Fecha)

# Ponemos estas fechas como indice
femicidios_indexados = femicidios.set_index(fechas)


# Creamos un dataframe vacio donde guardar ambas edades
edades = pd.DataFrame()

# La remuestreamos a años
femicidios_by_year = femicidios_indexados.resample('y')

# Nos quedamos con las edades promedio
edades['victima'] = femicidios_by_year.Edad.mean()
edades['victimario'] = femicidios_by_year.Edad_femicida.mean()

# Usemos el llamado de pandas a plot. No es lindo, pero es rápido.
edades.plot()
```
